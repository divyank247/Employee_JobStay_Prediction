```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hr_employee_churn_data.csv')
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df.describe()
```

```python
ax = sns.countplot(x='left',data =df)

for bars in ax.containers:
    ax.bar_label(bars)
#Takeaway : less people left
```

```python
ax = sns.countplot(x='salary',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
#Takeaway : Low salary people are more
```

```python
ax=sns.catplot(x='left',col ='salary',kind= 'count',data=df)
#Takeaway : low salary employee are having high leaving rate
```

```python
ax= sns.countplot(x = 'promotion_last_5years',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
#Takeaway : non promoting employee are very high
```

```python
sns.catplot(x='left', col = 'promotion_last_5years',kind='count',data =df)
#Takeaway : Non promoting employee leave more
```

```python
sns.countplot(x='Work_accident',data=df)
#takeaway : Less employees made accident
```

```python
sns.catplot(x ='left', col = 'Work_accident',kind ='count',data =df)
#Takeaway : the employee who made 0 accidents leave more. It seems this is a less imp feature.
```

```python
sns.displot(df['satisfaction_level'])
#Takeaway : sat_lev is distributed from 0.1 to 1 where it has high count at 0.1
```

```python
sns.boxplot(x= 'left',y='satisfaction_level',data = df)
#Takeaway : employee having sat_level between 0.2 to 0.5 left more
```

```python
sns.displot(df['number_project'])
#Takeaway : no. of project range from 2 to 7
```

```python
sns.boxplot(x='left',y='number_project',data = df)
#Takeaway : employees left when no. of project assigned were high
```
